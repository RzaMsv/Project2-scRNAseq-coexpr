```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
gene_exp_df = pd.read_csv("gene_expression_matrixP.csv", index_col=0)
gene_exp_transposed = gene_exp_df.transpose()
# gene_exp_transposed_NAN = gene_exp_transposed.replace(0, float('nan'), inplace=True)
# gene_exp_transposed_NAN = gene_exp_transposed.replace(1, float('nan'), inplace=True)
gene_exp_transposed.head()
```

```python
gene_exp_transposed['Pdcd1'].hist()

# set the labels and title of the histogram
plt.xlabel('Count')
plt.ylabel('Number')
plt.title('PD-1 expression')

# show the histogram
plt.show();
```

```python
num_max = gene_exp_transposed['Pdcd1'].max()
num_max
```

```python
# Define the gene you want to correlate with
target_gene = "Pdcd1"

# Get the column for the target gene
target_gene_col = gene_exp_transposed[target_gene]

# Calculate correlation between target gene and all other genes
corr_df = gene_exp_transposed.corrwith(target_gene_col)

# Sort correlations in descending order
corr_df = corr_df.sort_values(ascending=False)
corr_df.to_csv("correlations.csv")

# Print top correlations
print("Top correlations with {}:".format(target_gene))
print(corr_df.head(50))
```
